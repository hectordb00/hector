# 1. Line graph of weight and height

```python
data = { 'Names': ['Name1', 'Name2', 'Name3', 'Name4', 'Name5', 'Name6'],
        'Age': [25, 30, 30, 25, 25, 35],
        'Weight': [60, 62, 65, 63, 68, 70]
       }
df = pd.DataFrame(data)
df
```

```python
df.plot(kind = 'line')

plt.xticks(fontsize = 17)
plt.yticks(fontsize = 17)

plt.xlabel('x_label', fontsize = 17)
plt.ylabel('y_label', fontsize = 17)

plt.title('Comparison of Height and Weight of madeup data', fontsize = 17)

plt.show()
```

# 2. Represent `Male` and Female data of `CWDistance` in scatterplot

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.show()
```

# 3. Make a distribution of a categorical variable 'Gender' using swarmplot

```python
sns.swarmplot(x = 'Gender', 
           y = 'CWDistance',
           data = df
          )

plt.title('Swarplot of Gender', fontsize = 17)
plt.show()
```

### 4. What is the minimum, maximum, median, and quartiles of the variables , Height, Wingspan, CWDistance, CompleteGroup, Score using boxplot

```python
sns.boxplot(data = df.loc[:,['Height', 'Wingspan', 'CWDistance', 'CompleteGroup', 'Score']
                         ]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot', fontsize = 17)
plt.show()
```

### 5. Make a comparison b/w `Male`and `Female`using boxplot side by side

```python
sns.boxplot(data = df.loc[:,['Score']
                         ]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot', fontsize = 17)
plt.show()
```
