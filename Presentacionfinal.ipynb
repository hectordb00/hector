# 3. Who is more likely to get infected, men or women?

```python
sns.swarmplot(x = 'Sex', 
           y = 'Cases',
           data = df
          )

plt.title('Covid Cases between men and women', fontsize = 17)
plt.show()
```

With this swarplot I realized it does not matter if you're male or female, you can equally get inffected with Covid.

## 4. What age range (of men) had the most covid cases as of February 22, 2022?

```python
y = np.array([529, 1264, 10625, 2314])
mylabels = ["0 a 4 años", "10 a 14 años", "25 a 29 años", "70 a 74"]

plt.pie(y, labels = mylabels,autopct='%1.1f%%')
plt.show()
```

As seen visually in the pie chart the age range with the highest percentage of covid cases has to be those young men from ages 25 - 29 years old. This could be due to the fact that people around this age go out a lot.

## 5. How's the progression of Covid cases between men and women throughout time?

```python
sns.lmplot(x = 'Cases',
           y = 'Cases',
           data = df,
           fit_reg = False,
           hue = 'Sex'
          )
plt.show()
```

This scatter plot shows the progression of covid cases throught the days of men and women. As seen, both men and women have been equally infected without any exceptions of any kind.
