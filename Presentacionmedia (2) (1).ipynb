# Descripción de los datos

Descripción de las variables:
`Age Range ID`: Identificación de rango de edad de cuatro años

`Age range`: Rango de edad

`Sex ID`: Identificación de sexo (1=hombre, 2=mujer)

`Sex`: Género

`Time ID`: Identificación de Fecha

`Time`: Fecha

`Cases`: Casos confirmados ese día

`Total Cases`: Casos totales ese día

`Percentage`: Porcentaje de casos de los casos totales con los confirmados

# Preguntas

### Comparasión de casos covid el 22 de febrero del 2022 entre hombres y mujeres

```python
df = pd.read_excel('datasets/CasosCovid.xlsx',
                  engine = 'openpyxl',
                  )
df.head(37)
```

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.boxplot(data = df.loc[:,['Sex ID']
                         ]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot', fontsize = 17)
plt.show()
```

### ¿Cúal es la mínima, máxima, media, y cuartitle de las variables `Cases` y `Total Cases`?

```python
sns.boxplot(data = df.loc[:,['Cases', 'Total Cases']
                         ]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot', fontsize = 17)
plt.show()
```

# ¿Quién se contagia más, hombres o mujeres?

```python
sns.swarmplot(x = 'Sex', 
           y = 'Cases',
           data = df
          )

plt.title('Casos covid entre hombres y mujeres', fontsize = 17)
plt.show()
```
