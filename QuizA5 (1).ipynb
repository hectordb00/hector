```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
df = pd.read_csv('datasets/A5.csv')
df
```

```python
sns.boxplot(data = df.loc[:,['peso', 'sexo', 'edad', 'cintura']
                         ]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot', fontsize = 17)
plt.show()
```

```python
sns.swarmplot(x = 'peso', 
           y = 'cintura',
           data = df
          )

plt.title('peso y cintura', fontsize = 17)
plt.show()
```

```python
counts = df['sexo'].value_counts()
counts = counts.to_frame()
type(counts)
```

```python
counts['%'] = (counts.iloc[:, :]/counts.loc[:, 'sexo'].sum(axis=0))*100
counts 
```

```python
sns.lmplot(x = 'peso',
           y = 'cintura',
           data = df,
          )
plt.show()
```
